## Prepare Dataset

```python
def get_min(time_str, am_or_pm):
    """Get minutes from time."""
    h, m = int(time_str.split(':')[0]), int(time_str.split(':')[1])
    if am_or_pm == "PM":
        h += 12
    return 60*h + m

int_day_df["time"] = int_day_df["TIME"].apply(lambda l : get_min(l.split(" ")[1], l.split(" ")[2]))
```

```python
int_day_df.tail()
```

```python
arrival_times = []
departure_times = []

for transaction, time in zip(int_day_df["TRANSACTION"], int_day_df["time"]):
    if transaction == "DEP":
        arr_time = time - random.randint(40, 120)
        departure_times.append(time)
        arrival_times.append(int(arr_time))
    if transaction == "ARR":
        dep_time = time - random.uniform(40, 120)
        departure_times.append(int(dep_time))
        arrival_times.append(time)
int_day_df["departure_time"] = departure_times
int_day_df["arrival_time"] = arrival_times
```

```python
int_day_df
```

```python
arrivals_df = int_day_df[int_day_df["TRANSACTION"] == "ARR"][["FLIGHT_NUMBER", "departure_time", "arrival_time"]]
arrivals_df.to_csv("arrivals.csv")
```

```python
departures_df = int_day_df[int_day_df["TRANSACTION"] == "DEP"][["FLIGHT_NUMBER", "departure_time", "arrival_time"]]
departures_df.to_csv("departures.csv")
```
